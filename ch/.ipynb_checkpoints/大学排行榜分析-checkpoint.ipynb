# 数据处理实战案例（2）：大学排行榜分析

## 题目：从软科大学排名榜获取2015~2021年中国大学排名，分析我校的情况

### 提示：
#### （1）使用requests获取网页数据 
  #### （url = 'https://www.shanghairanking.cn/api/pub/v1/bcur?bcur_type=11&year=2021'）  
  #### 替换年份可以得到2015~2021年的排名JSON格式数据

```python
import requests
import numpy as np
import pandas as pd
np.random.seed(12345)
import matplotlib.pyplot as plt
ccnu_data=[]
```

```python
for i in range(2015,2021):
    
    s = 'https://www.shanghairanking.cn/api/pub/v1/bcur?bcur_type=11&year='+str(i)
    url = s
    resp = requests.get(url)
    data = resp.json()
    ranking= pd.DataFrame(data['data'],columns=['rankings'])
    uni_data=ranking.values
    
    find=1
    count=0
    while find:
        if uni_data[count][0]['univNameCn']=='华中师范大学':
            ccnu_data.append(uni_data[count][0])
            find=0
        count=count+1
ccnu_data
```

```python
#排名
rank=[]
for i in range(0,6):
    rank.append(ccnu_data[i]['ranking'])
rank
    
```
