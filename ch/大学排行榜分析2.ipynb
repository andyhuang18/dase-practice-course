# 数据处理实战案例（2）：大学排行榜分析

## 题目：从软科大学排名榜获取2015~2021年中国大学排名，分析我校的情况

### 提示：
#### （1）使用requests获取网页数据 
  #### （url = 'https://www.shanghairanking.cn/api/pub/v1/bcur?bcur_type=11&year=2021'）  
  #### 替换年份可以得到2015~2021年的排名JSON格式数据

```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
plt.rc('figure', figsize=(20, 10))
plt.rc('lines', lw=2, color='red')

import requests
from bs4 import BeautifulSoup
import json
```

```python
def request_data(year):
    try:
        if(year >2014 and year <2022):
            url = 'https://www.shanghairanking.cn/api/pub/v1/bcur?bcur_type=11&year=%d'%year
            response = requests.get(url, timeout=20)
            data = response.json()['data']['rankings']
            inds_name = response.json()['data']['inds']
            
            return data,inds_name
        else:
            return 'no data','no inds_name'
    except:
        return 'no data','no inds_name'
```

```python
def  flatten_data(data):
    flat_data = []
    
    for d in data:
        indData = d['indData']
        del d['indData']
        d.update(indData)
        flat_data.append(d)

    return flat_data
        
```

```python
def parse_indsName(inds_name):
    inds_dict={}
    
    for index in inds_name:
        inds_dict[index['code']] = index['nameCn']

    return inds_dict
```

```python
def parse_cols(inds_dict):
    ext_cols    = ['ranking','univNameCn','univTags','univCategory','province','score']
    ext_col_sh  = ['排名', '学校名称', '标签','类别','省份','得分']
    inds_col    = []
    inds_col_sh = []

    for ind in inds_dict:
        inds_col.append(ind)
        inds_col_sh.append(inds_dict[ind])
        
    ext_cols.extend(inds_col)
    ext_col_sh.extend(inds_col_sh)

    return ext_cols,ext_col_sh
```

```python
def parse_to_tb(year):
    data,inds_name = request_data(year)
    flat_data = flatten_data(data)
    inds_dict = parse_indsName(inds_name)
    ext_cols, ext_col_sh = parse_cols(inds_dict)
    
    rank_table = pd.DataFrame(flat_data, columns = ext_cols)
    rank_table.columns = ext_col_sh
    
    rank_table['年度'] = year
    rank_table['排名'] = rank_table['排名'].astype('int')
    
    return rank_table
```

```python
rank_tbs = []

for year in range(2015, 2022):
    rank_tbs.append(parse_to_tb(year))
```
